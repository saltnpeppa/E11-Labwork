```python
plt.hist(pressure, bins = 15)
plt.ylabel('Frequency')
plt.xlabel('Pressure in Pascals')
plt.title("Pressure Vs. Time")
mean = np.mean(pressure)
sd = np.std(pressure)
x = np.linspace(mean - 4*sd, mean + 4*sd, 100)
plt.plot(x, norm.pdf(x, mean, sd))

plt.show()
```

```python
plt.hist(relative_humidity, bins = 15)
plt.ylabel('Frequency')
plt.xlabel('Percent Relative Humidity')
mean = np.mean(relative_humidity)
sd = np.std(relative_humidity)
x = np.linspace(mean - 4*sd, mean + 4*sd, 100)
plt.plot(x, norm.pdf(x, mean, sd) * mean)


plt.show()
```
